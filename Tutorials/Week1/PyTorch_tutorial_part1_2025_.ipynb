# Why PyTorch
Pytorch is important for deep learning because it makes building, training, and experimenting with neural networks easy and fast.

# Definition


PyTorch is a Python-based library used to build neural networks.

It provides two main features:

* An n-dimensional Tensor object

* Automatic differentiation for building and training neural networks


## Tensors

At its core, PyTorch is a library for processing tensors. Tensors are a specialized data structure similar to NumPy's ndarrays, except that tensors can run on GPUs (to accelerate their numeric computations). This is a major advantage of using tensors. A tensor can be a number, vector, matrix, or any n-dimensional array.


![picture](https://tse3.mm.bing.net/th/id/OIP.gq-qHnpwhiReloHgqP6sdQHaCl?rs=1&pid=ImgDetMain&o=7&rm=3)

PyTorch supports multiple types of tensors, including:

1. FloatTensor: stores numbers as 32-bit floating-point values.

2. DoubleTensor: stores numbers as 64-bit floating-point values (higher precision).

3. HalfTensor: stores numbers as 16-bit floating-point values (lower precision).

4. IntTensor: stores numbers as 32-bit signed integers.

5. LongTensor: stores numbers as 64-bit signed integers.

# Torch tensors and numpy conversion

### From tensor to numpy

```python
## Tensor to numpy array
a = torch.rand(3,3)
print(a, '\n')

## convert a into numpy array b
b = a.numpy()
print(b, '\n')

## update the tensor a by adding a value to it
b +=2
print(a, '\n')
```

Note: if the tensor a is on the cpu and not the gpu, then both a and b will share/point to the same memory

### From numpy to tensor

```python
# torch.set_printoptions(precision=8)
import numpy as np

a = np.random.rand(3,3)
print(a, '\n')

## convert array a into tensor b
b = torch.from_numpy(a)#.clone()
print(b, '\n')
a += 2
print(b)
```

### Move tensor to a specific  GPU/CPU device

```python
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
a = torch.ones(2, device=device)
print(a)
```

```python
#Move to GPU if available
if torch.cuda.is_available():
  device = torch.device('cuda')
  print(device)

  x = torch.ones(5, device=device)
  y = torch.ones(5).to(device)
  z = x+y

  z.numpy()
```

Note: if we want to convert the z below directly to numpy (z.numpy()), this will return an error, because numpy can only handle cpu tensors; So we wil have to bring first z to cpu first and later on do the convertion

```python
# z.to("cpu").numpy()
z.cpu().numpy()
```

# Exercise 3:

Create a function that takes in a square matrix A and returns a 2D tensor consisting of a flattened A with the index of each element appended to this tensor in the row dimension, e.g.,



 \begin{equation*}
 A=
 \begin{bmatrix}
  2 & 3 \\
  4 & -2
\end{bmatrix}, \ \
output = \begin{bmatrix}
  0 & 2 \\
  1 & 3 \\
  2 & 4 \\
  3 & -2
\end{bmatrix}
\end{equation*}


```python
def function_concat(A):
    # Flatten the input tensor A
    flattened_A = A.flatten()

    # Create a tensor of indices from 0 to num_elements-1
    # Ensure the dtype matches that of A for consistent stacking
    indices = torch.arange(flattened_A.numel(), dtype=A.dtype)

    # Stack the indices and the flattened tensor column-wise
    # The result will be a 2D tensor where the first column is indices and the second is values
    output = torch.stack((indices, flattened_A), dim=1)
    return output

## TEST YOUR CODE HERE ##
A = torch.tensor([[2,3],[4,-2]])
print(function_concat(A))
```
